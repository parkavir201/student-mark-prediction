```python
import pandas as pd

# dataset create
data = {
    "Hours_Study": [1,2,3,4,5,6,7,8,2,3,6,7],
    "Attendance": [50,60,65,70,75,80,85,90,55,68,78,88],
    "Marks": [30,35,40,50,55,65,70,80,33,45,60,75]
}

df = pd.DataFrame(data)

# input output
X = df[["Hours_Study","Attendance"]]
y = df["Marks"]

# model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

# prediction
import pandas as pd

sample = pd.DataFrame([[5,80]], columns=["Hours_Study","Attendance"])
model.predict(sample)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Hours_Study"], df["Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X)
print("Accuracy (R2 Score):", r2_score(y, y_pred))
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Hours_Study"], df["Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study vs Marks")
plt.show()
```
